```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/User/Desktop/New folder (2)/Python/Project 1  Supermarket sale/Data/supermarket_sales.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month_name()
```

```python
total_revenue = df['Total'].sum()
total_revenue
```

```python
monthly_sales = df.groupby('Month')['Total'].sum()
monthly_sales
```

```python
monthly_sales.plot(kind='line',title='Monthly sales Trend')
plt.show()
```

```python
df.groupby('Customer type')['Total'].mean()
```

```python
product_sales = df.groupby('Product line')['Total'].sum()

product_sales.plot(kind='bar', title='Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```
